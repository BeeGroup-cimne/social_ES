# Population Census

The [Annual Population Census](https://www.ine.es/dynt3/inebase/en/index.htm?padre=11555&capsel=11100) provides detailed demographic statistics for all census sections in Spain, identified by country, province, municipality, district, and section codes, along with the reference year. It includes total population counts disaggregated by nationality, sex, birth origin, and birth country, covering both Spanish and foreign residents across multiple specific nationalities. Data also capture birthplace distribution within Spain (same municipality, province, autonomous community, or other regions) and abroad. Age structure is available in five-year intervals, further broken down by sex. The dataset offers highly granular insights into the composition and spatial distribution of the population, enabling analyses of migration patterns, gender distribution, and demographic ageing. Its fine territorial resolution supports local planning, socio-economic studies, and policy-making, particularly for municipalities, districts, and neighbourhoods.


```python
population_df = INE.PopulationCensus(wd=wd)
```

These are the geographical levels available:

```python
pprint.pp(list(population_df.keys()))
```

An example of the results for the municipality of Barcelona at census tract level is the following:

```python
population_df["Sections"][population_df["Sections"]["Municipality code"]=="08019"]
```

The columns available in this dataset are:

```python
pprint.pp(list(population_df['Sections'].columns), width=400)
```

# Population Education and Employment Status Census

This [dataset](https://www.ine.es/dynt3/inebase/en/index.htm?padre=10608&capsel=10613) provides detailed information on the educational attainment and labour market situation of the population in Spain, disaggregated to the census section level and identified by country, province, municipality, district, and section codes, as well as reference year. Educational attainment is classified into four categories: primary or below, lower secondary, upper secondary and post-secondary non-tertiary, and tertiary education. Data are further broken down by sex, place of birth (Spain or abroad), and their intersections. Labour force status includes employed, unemployed, recipients of disability or retirement pensions, other inactive situations, and students. These variables are also cross-classified by sex and place of birth, enabling nuanced analysis of participation, unemployment, and education levels across different population groups. The dataset supports local and regional assessments of human capital, workforce composition, and socio-economic inequalities, making it a valuable resource for policy design, education planning, and labour market studies.

# Households Price Index

This [dataset](https://www.ine.es/dyngs/INEbase/en/operacion.htm?c=Estadistica_C&cid=1254736152838&menu=ultiDatos&idp=1254735976607) provides quarterly indicators of housing prices in Spain, broken down by autonomous community and province, identified through corresponding territorial codes and names, along with the reference year and quarter. It distinguishes between three segments: first-hand housing market (new dwellings), second-hand housing market (previously owned dwellings), and the whole housing market (combined). The indices measure price evolution over time, enabling analysis of regional and national trends in residential property values. This information is essential for monitoring real estate market dynamics, assessing housing affordability, and evaluating the impact of economic cycles and policy measures on the housing sector. The dataset supports applications in urban planning, investment analysis, and socio-economic studies related to housing supply and demand.

```python
housing_market_df = INE.HouseholdsPriceIndex(wd=wd)
```

# Essential Characteristics of Population and Households by building

This [dataset](https://www.ine.es/dyngs/INEbase/es/operacion.htm?c=Estadistica_C&cid=1254736177092&menu=resultados&idp=1254735572981#_tabs-1254736195788) provides highly detailed socio-demographic and housing information for the population residing in each building in Spain, obtained by extrapolating survey results to the building level. It includes characteristics of residents aged 16 and over, such as workplace/study location, commuting patterns, transport mode, domestic task participation, caregiving responsibilities, social support, digital access, and online purchasing habits. It records household composition (e.g., single-person households, cohabiting couples, number of children), housing attributes (size, number of rooms, ownership/rental regime, mortgage or rent costs, presence of a second residence), and asset ownership (vehicles, eco-friendly vehicles). Environmental and energy-related data include waste separation practices, type of heating and fuel, cooling systems, internet connection, appliances, lighting, and renewable energy installations. The dataset also captures building condition, accessibility, surrounding area problems, and availability of infrastructure or services. This rich granularity supports analyses of living conditions, housing quality, socio-economic diversity, and environmental practices at a micro-geographic scale.
